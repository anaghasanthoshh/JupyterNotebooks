```python
df=pd.read_csv('tips.csv')
```

```python
df.head()
```

```python
def quality(total_bill,tip):
    val=tip/total_bill
    if val>0.25:
        return 'Generous'
    else:
        return 'Other'
    
```

```python
df['Quality']=df[['total_bill','tip']].apply(lambda df:quality(df['total_bill'],df['tip']),axis=1)


```

```python
df.head()
```

```python
#Using Vectorize
df['Vectorised_qual']=np.vectorize(quality)(df['total_bill'],df['tip'])
```
