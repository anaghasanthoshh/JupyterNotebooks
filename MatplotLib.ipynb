```python
from sklearn.datasets import make_blobs
X, y = make_blobs(100, 2, centers=2, random_state=2, cluster_std=1.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdGy');
#Here c is a third variable, and the colours either represent a specif group or number range
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)  # The third variable

# Create scatter plot with a colormap
plt.scatter(x, y, c=z, cmap='viridis')
plt.colorbar()  # Add a color bar to show the mapping
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Colormap')
plt.show()
#here since everything is random,we dont see a pattern when coloring wrt to z
```

```python
#Barplots
x1=[1,5]
y1=[10,2]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x2,y2,color='r')
plt.bar(x1,y1,color='y')
```

```python
#Piechart
#Needs 4 params-sizes,labels,colors,explode
#size-is the values,colors-respective color u want,labels-the names,
#explode-which one to be projected out

labels=['apples','oranges','bananas','strawberry']
colors=['red','orange','yellow','pink']
explode=(0,0,0,0.1)
size=[30,10,5,80]

plt.pie(size,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)

```
