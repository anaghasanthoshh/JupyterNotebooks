```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
#scatter plot:
plt.scatter(x,y,c='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter plot')

plt.savefig('file.png')
```

```python
y=x*x
```

```python
plt.plot(x,y,'b--')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
#Subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

```python
#Barplots
x1=[1,5]
y1=[10,2]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x2,y2,color='r')
plt.bar(x1,y1,color='y')
```

```python
#Piechart
#Needs 4 params-sizes,labels,colors,explode
#size-is the values,colors-respective color u want,labels-the names,
#explode-which one to be projected out

labels=['apples','oranges','bananas','strawberry']
colors=['red','orange','yellow','pink']
explode=(0,0,0,0.1)
size=[30,10,5,80]

plt.pie(size,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)

```
